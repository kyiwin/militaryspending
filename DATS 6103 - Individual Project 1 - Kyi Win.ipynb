```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# import the military spending dataset
dfmsud = pd.read_csv('msud.csv', index_col = 0)
```

```python
# scan the dataset
dfmsud.head(10)
```

```python
# take out data of 20 countries with Highest GDP and 11 years from 2006 to 2016
msud = dfmsud.loc[['USA','UK','Germany','Russia','France','Italy','India','Brazil','Spain','Canada','Australia', 'Mexico','China', 'South Korea','Japan','Indonesia','Turkey','Netherlands','Switzerland','Saudi Arabia'],['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']]
msud
```

```python
# check the data type
msud.dtypes
```

```python
# change the data type
cols =['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']
msud[cols]= msud[cols].apply(pd.to_numeric, errors = 'coerce', axis = 1)
msud.dtypes
```

```python
# exploring the data set of military spending 
# graphing Military spending of 20 countries in US$ over 11 years
msudp= msud.T.plot(figsize=(20,10), fontsize=13, legend=True)
plt.legend(loc='best', fontsize=13)
msudp.set_xlabel('Year')
msudp.set_ylabel('US dollar in million')
msudp.set_title('Military Spending US Dollar')
plt.show()
```

```python
# summary of the dataset
msud.describe()
```

```python
# look for the amount that each country spend the most in military spending
msud['Max Value'] = msud.max (axis = 1)
msud['Max Value']
```

```python
# rank them by the order 
msud.sort_values(by= 'Max Value', ascending = False)
```

```python
# look for the year that they spend the most in military spending
msud['Highest_Year'] = msud[['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']].idxmax(axis=1)
msud['Highest_Year']
```

```python
# look for minimum value the countires spent for military spending over 11 years
msud['Min Value'] = msud.min(axis = 1)
msud['Min Value']
```

```python
# rank them by minimum value 
msud.sort_values(by= 'Min Value', ascending = True)
```

```python
# look for the year that they spent the lowest
msud['Lowest_Year'] = msud[['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']].idxmin(axis=1)
msud['Lowest_Year']
```

```python
# overall data become
msud
```

```python
# take out max and min value to graph
comvalue= msud.loc[:,['Max Value', 'Min Value']]
comvalue
```

```python
# graphing the min and max value 
comvalue.plot(kind = 'barh', sort_columns = True, figsize = (10,5))
plt.xlabel('Military Spending in Million')
plt.show()
```

```python
# deleting four columns
del msud['Max Value']
del msud['Highest_Year']
del msud['Min Value']
del msud['Lowest_Year']
```

```python
# 1.import the GDP dataset to compare the data to that country's GDP
dfgdp = pd.read_csv('GDP.csv', error_bad_lines = False, index_col = 0)
```

```python
# scan the dataset
dfgdp
```

```python
# take out % from the dataset for calculation
dfgdp[cols] = dfgdp[cols].replace({'%': ''}, regex = True)
dfgdp
```

```python
# take out 11 years from 2006 to 2016 and 20 countries with highest GDP
gdp = dfgdp.loc[['USA','UK','Germany','Russia','France','Italy','India','Brazil','Spain','Canada','Australia', 'Mexico','China', 'South Korea','Japan','Indonesia','Turkey','Netherlands','Switzerland','Saudi Arabia'],['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']]
gdp
```

```python
# change the data type to float
cols =['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']
#msud[cols]= msud[cols].apply(pd.to_numeric, errors = 'coerce', axis = 1)
gdp[cols]= gdp[cols].apply(pd.to_numeric, errors = 'coerce', axis = 1)
```

```python
# check the data type
gdp.dtypes
```

```python
# transform GDP to calculate shared GDP %
gdpt = gdp.T
gdpt
```

```python
# plot military spending of shared GDP %
sgdp= gdp.T.plot(figsize=(20,10), fontsize=13, legend=True)
plt.legend(loc='best', fontsize=13)
sgdp.set_xlabel('Year')
sgdp.set_ylabel('% shared gdp')
sgdp.set_title('Military Spending by % Shared gdp by years')
plt.show()
```

```python
# the country's total gdp calculation
totalgdp = (msud*100) / (gdp)
totalgdp
```

```python
# plot the total GDP
tgdpp= totalgdp.T.plot(figsize=(20,10), fontsize=13, legend=True)
plt.legend(loc='best', fontsize=13)
tgdpp.set_xlabel('Year')
tgdpp.set_ylabel('GDP in million')
tgdpp.set_title('Total GDP')
plt.show()
```
